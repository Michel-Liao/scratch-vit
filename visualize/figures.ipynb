```python
import matplotlib.pyplot as plt
import numpy as np

# Data
patch_sizes = [1, 2, 4, 7]
learning_rates = [-4, -5, -6, -7]
accuracies = [
    [0.9426, 0.9350, 0.8684, 0.8470],
    [0.9122, 0.8796, 0.7894, 0.8216],
    [0.7248, 0.6384, 0.5162, 0.6670],
    [0.1446, 0.2356, 0.1796, 0.2732]
]

# Create a grid for patch_size and learning_rate
X, Y = np.meshgrid(patch_sizes, learning_rates)

# Flatten data for plotting
sizes = np.exp(np.array(accuracies).flatten() * 7.5) + 250  # Scale accuracy to circle size
colors = np.array(accuracies).flatten()  # Use accuracy for color mapping

# Create scatter plot
plt.figure(figsize=(8, 8))
scatter = plt.scatter(
    X.flatten(), Y.flatten(), s=sizes, c=colors, cmap="PiYG", alpha=1, edgecolors="k"
)

# Add a colorbar to represent the accuracy
cbar = plt.colorbar(scatter)
cbar.set_label("Accuracy", fontsize=20)

# Labels and title
plt.xlabel("Patch Size (px)", fontsize=20)
plt.ylabel("Learning Rate (10^x)", fontsize=20)
plt.xlim(0, 8)
plt.ylim(-7.5, -3.5)
plt.xticks(patch_sizes)
plt.yticks(learning_rates)
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

```python
# Data extracted from the image
epochs = np.arange(1, 8)
training_loss = [0.5862, 0.1470, 0.0696, 0.0425, 0.0266, 0.0257, 0.0347]
test_loss = [0.73, 0.1742, 0.169, 0.1686, 0.193, 0.2399, 0.22]

# Plotting
plt.figure(figsize=(10, 6))

# Plot training loss
plt.plot(epochs, training_loss, label='Training Loss', marker='o', linestyle='-', color='blue')

# Plot test loss (ignoring None values)
test_epochs = [e for e, t in zip(epochs, test_loss) if t is not None]
test_loss_clean = [t for t in test_loss if t is not None]
plt.plot(test_epochs, test_loss_clean, label='Test Loss', marker='s', linestyle='--', color='orange')

# Adding details
plt.title('Training and Test Loss over Epochs', fontsize=20)
plt.xlabel('Epoch', fontsize=20)
plt.ylabel('Loss', fontsize=20)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.xticks(epochs)

# Show plot
plt.tight_layout()
plt.show()
```
